```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the image
image = cv2.imread("images/lena.png")

# Resize the image to a fixed width and height
#resized_image = cv2.resize(image, (800, 1200))
resized_image = cv2.resize(image, None, fx=1.75, fy=1.75)

#show the resized image

cv2.imshow("Resized Image", resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
