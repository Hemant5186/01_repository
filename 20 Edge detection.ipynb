```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('images/lenna.png', 0)
```

```python
# Sobel parameters: ksize, dx, dy, ddepth
# Sobel for horizontal edges
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)

# Sobel for vertical edges
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

# Combine the sobel images (optional)
combined_sobel = cv2.add(sobelx, sobely)
```

```python
final_img = np.hstack([sobelx, sobely, combined_sobel])
cv2.imshow('All images', final_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,4,1);
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.subplot(1,4,2)
plt.title('Sobel x')
plt.imshow(sobelx, cmap='gray')
plt.subplot(1,4,3)
plt.title('Sobel y')
plt.imshow(sobely, cmap='gray')
plt.subplot(1,4,4)
plt.title('Final')
plt.imshow(combined_sobel, cmap='gray')
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,4,1);
plt.title('Original Image')
plt.imshow(abs(img), cmap='gray')
plt.subplot(1,4,2)
plt.title('Sobel x')
plt.imshow(abs(sobelx), cmap='gray')
plt.subplot(1,4,3)
plt.title('Sobel y')
plt.imshow(abs(sobely), cmap='gray')
plt.subplot(1,4,4)
plt.title('Final')
plt.imshow(abs(combined_sobel), cmap='gray')
```

##### canny edge detection 

```python
import cv2
img = cv2.imread('images/circle.png', 0) # 0 for load as gray scale

# Apply canny edge detection

edges = cv2.Canny(img, 50, 150) # Adjust thresholds if needed

# Display results
cv2.imshow('Original images', img)
cv2.imshow('Canny edge detection images', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()

# or 
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Original Image')
plt.imshow(edges)
```
