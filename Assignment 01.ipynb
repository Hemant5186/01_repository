```python
video = cv2.VideoCapture('images/nature.mp4')
```

```python
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
fps=int(video.get(cv2.CAP_PROP_FPS))
```

```python
print("Video Resolution:", height, width)
print(" Total frame:", frames)
print("fps", fps)
```

```python
# 4 b
while True:
    ret,frame = video.read()
    if not ret:
        break
        
    cv2.imshow("video", frame)
    if cv2.waitKey(15) ==ord('q'):
        break
cv2.destroyAllWindows()
```

```python
timestamp='00:00:50'
timestamp.split(':')
```

```python
hours,minutes,seconds=[float(i) for i in timestamp.split(':')]
print(hours,minutes,seconds)
```

```python
trigger_frame = hours*3600*fps + minutes*60*fps
```

```python
video.set(1,trigger_frame)
sucess, frame = video.read()
if sucess:
    cv2.imwrite('sc20.jpg', frame)
```
