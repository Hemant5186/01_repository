```python
# create tensors
a = tf.constant([1,2,3])
b = tf.constant([4,5,6])

print("Tensor a:", a)
print("Tensor b:", b)
```

```python
# create tensors
x = tf.constant([[1,2,3],[7,8,9]], dtype=np.int8)
y = tf.constant([[4,5], [6,2]])

print("Tensor a:", x)
print("Tensor b:", y)
```

```python
# Multiply tensors
e = tf.multiply(a,b)
print("Multiplication:", e)
```

```python
# Divide tensors
f = tf.divide(a, b)
print("Division:", f)
```

```python
# Matrix multiplication
a_matrix = tf.constant([[1,2], [3,4]])
b_matrix = tf.constant([[5,6], [7,8]])

g = tf.matmul(a_matrix, b_matrix)
print("Matrix Multiplication:", g)
```
