```python
#8
bowler_max_matches = df[df['wickets'] > 0].sort_values(by = 'matches', ascending=False).iloc[0]
```

```python
print('Bowler who played highest number of matches:', bowler_max_matches['name'])
```

```python
#9
average_wickets_bowlers = df[df['wickets'] > 0]['wickets'].mean()
print(average_wickets_bowlers)
```

```python
#10
df['bowling_average'] = df['wickets'] / df['matches']
bowler_least_average = df[df['wickets'] > 0].sort_values(by = 'bowling_average').iloc[0]
print(bowler_least_average)
```

```python
#11
df['catches_per_match'] = df['catches'] / df['matches']
catches_ratio_above_0_5 = df[df['catches_per_match'] > 0.5]
print(catches_ratio_above_0_5)
```

```python
#12
plt.scatter(df['matches'], df['runs'])
plt.xlabel('matches')
plt.ylabel('runs')
plt.title('Matches vs Runs')
plt.show()
```

```python
# 13
sorted_by_runs = df.sort_values(by = 'runs')
print(sorted_by_runs)
```

```python
# 14
wicket_greater_than_matches = df[df['wickets'] > df['matches']]['name']
```
