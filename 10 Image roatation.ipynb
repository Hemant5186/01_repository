##### Roation using function

```python
import cv2
# load the image
image = cv2.imread("images/lena.png")

# Roatate the image 90 degrees clockwise
rotated_image = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)

# Show the rotated  images
cv2.imshow("Rotated Image", rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Create the rotation matrix at angle 60

```python
# load the image
image = cv2.imread("images/lena.png")

# Define the center point and rotation angle
center = (image.shape[1] // 2, image.shape[0] // 2)
angle = -60

# Create the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)

# Apply the rotation using warpAffine
rotated_image = cv2.warpAffine(image, rotation_matrix,(image.shape[1], image.shape[0]))

# show the rotated image
cv2.imshow("Rotated Image", rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
