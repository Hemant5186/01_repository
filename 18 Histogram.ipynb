```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num = [np.random.randint(1,100) for x in range(1000)]
```

```python
num
```

```python
print(num)
```

```python
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Count')
plt.xticks(range(10,101,10))
plt.grid()
plt.hist(num, bins=range(10,101,10))
```
