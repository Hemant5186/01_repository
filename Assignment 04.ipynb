```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the image
image = cv2.imread('images/sachin.jpg', 0)
image1 = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image1.shape

#cv2.imshow('image', image)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
#image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
#image.shape
#plt.imshow(image)
```

```python
#1
height, width =image.shape[:2]
T = np.float32([[1,0,400],[0,1,-150]])
img_translation = cv2.warpAffine(image, T, (width,height))
```

```python
cv2.imshow('Original Image', image)
cv2.imshow('Translated Image', img_translation)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 2
center=(100,100)
angle=-60
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
rotated_image = cv2.warpAffine(image, rotation_matrix,(image.shape[1],image.shape[0]))
cv2.imshow("Rotated Image", rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 3
resized_image = cv2.resize(image1,None, fx=0.75, fy=0.75)
```

```python
plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(image1)
plt.subplot(1,2,2)
plt.title('Resized')
plt.imshow(resized_image)
```

```python
#4
flipped_hr = cv2.flip(image1, 0)
flipped_vr = cv2.flip(image1, 1)
flipped_dg = cv2.flip(image1, -1)

plt.figure(figsize=(16,9))
plt.subplot(1,4,1)
plt.title('Original Image')
plt.imshow(image1)
plt.subplot(1,4,2)
plt.title('Horizontal flipped Image')
plt.imshow(flipped_hr)
plt.subplot(1,4,3)
plt.title('Vertical flipped  Image')
plt.imshow(flipped_vr)
plt.subplot(1,4,4)
plt.title('Diagonal flipped Image')
plt.imshow(flipped_dg)
```

```python
#5
cropped_image = image1[300:500,400:560]
plt.subplot(1,2,1)
plt.imshow(image1)
plt.subplot(1,2,2)
plt.imshow(cropped_image)
```
