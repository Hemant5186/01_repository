```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
image = cv2.imread('images/lena.png')
```

```python
# Define a 3x3 averging kernal
kernal = np.ones((7, 7), np.float32) / 49
```

```python
# Apply the kernal to smooth the image
```

```python
smoothed_image = cv2.filter2D(image, -1, kernal)
```

```python
# show the original and smoothed images
cv2.imshow('Original', image)
cv2.imshow('Smoothed', smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
